```python
import tkinter as tk
from tkinter import messagebox, simpledialog
import cv2
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import random
import os
from datetime import datetime


FROM_EMAIL = "rishimopuri@gmail.com"
FROM_PASSWORD = "eobm egvo tpyy oqkg"  


def send_email(to_email, subject, body):
    msg = MIMEMultipart()
    msg['From'] = FROM_EMAIL
    msg['To'] = to_email
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(FROM_EMAIL, FROM_PASSWORD)
        server.sendmail(FROM_EMAIL, to_email, msg.as_string())
        server.quit()
    except Exception as e:
        print(f"Email error: {e}")
        messagebox.showerror("Email Error", f"Could not send email.\n{e}")

# User File Handling 
def load_users():
    users = {}
    try:
        with open("users.txt", "r") as file:
            for line in file:
                parts = line.strip().split(",")
                if len(parts) != 3:
                    continue
                username, email, password = parts
                users[username] = (email, password)
    except FileNotFoundError:
        pass
    return users

def save_user(username, email, password):
    users = load_users()
    if username in users:
        return False
    with open("users.txt", "a") as file:
        file.write(f"{username},{email},{password}\n")
    return True

# Camera Capture 
def capture_image(username):
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    if ret:
        if not os.path.exists("faces"):
            os.makedirs("faces")
        cv2.imwrite(f"faces/{username}.jpg", frame)
        cap.release()
        messagebox.showinfo("Camera", "Face image captured successfully!")
    else:
        messagebox.showerror("Camera", "Failed to capture image.")
    cap.release()

# login 
class LoginApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Smart Automation")
        self.root.geometry("400x300")
        self.root.resizable(False, False)

        self.username = tk.StringVar()
        self.password = tk.StringVar()

        tk.Label(root, text="Smart Automation Portal", font=("Arial", 16, "bold")).pack(pady=10)

        tk.Label(root, text="Username").pack()
        tk.Entry(root, textvariable=self.username).pack()

        tk.Label(root, text="Password").pack()
        tk.Entry(root, textvariable=self.password, show="*").pack()

        tk.Button(root, text="Login", command=self.login).pack(pady=5)
        tk.Button(root, text="Sign Up", command=self.open_signup_window).pack()

    def login(self):
        users = load_users()
        username = self.username.get()
        if username in users:
            email, password = users[username]
            if password == self.password.get():
                messagebox.showinfo("Login", f"Welcome, {username}!")
                self.open_dashboard(username)
            else:
                messagebox.showerror("Login Failed", "Incorrect password.")
        else:
            messagebox.showerror("Login Failed", "Username not found.")

    def open_signup_window(self):
        SignupWindow(self.root)

    def open_dashboard(self, username):
        dashboard = tk.Toplevel(self.root)
        dashboard.title("Dashboard")
        dashboard.geometry("400x300")
        tk.Label(dashboard, text=f"Welcome, {username}", font=("Arial", 14)).pack(pady=10)
        tk.Button(dashboard, text="Capture Face Again", command=lambda: capture_image(username)).pack(pady=10)

#Signup Process 
class SignupWindow:
    def __init__(self, master):
        self.top = tk.Toplevel(master)
        self.top.title("Email Verification")
        self.top.geometry("400x200")
        self.email = tk.StringVar()
        self.code = None

        tk.Label(self.top, text="Enter your email to start signup").pack(pady=10)
        tk.Entry(self.top, textvariable=self.email, width=40).pack(pady=5)
        tk.Button(self.top, text="Send Verification Code", command=self.send_verification_code).pack(pady=5)

    def send_verification_code(self):
        self.code = random.randint(100000, 999999)
        email = self.email.get()
        send_email(email, "Verification Code", f"Your verification code is: {self.code}")
        messagebox.showinfo("Email Sent", "Verification code sent to your email.")
        self.ask_verification_code()

    def ask_verification_code(self):
        entered_code = simpledialog.askinteger("Verification", "Enter the code sent to your email:")
        if entered_code == self.code:
            messagebox.showinfo("Verified", "Email verified successfully!")
            self.ask_credentials()
        else:
            messagebox.showerror("Error", "Incorrect verification code.")

    def ask_credentials(self):
        username = simpledialog.askstring("Username", "Choose a username:")
        password = simpledialog.askstring("Password", "Choose a password:", show='*')
        email = self.email.get()
        if save_user(username, email, password):
            capture_image(username)
            messagebox.showinfo("Sign Up", "Signup successful! Please login now.")
            self.top.destroy()
        else:
            messagebox.showerror("Error", "Username already exists.")


if __name__ == "__main__":
    root = tk.Tk()
    app = LoginApp(root)
    root.mainloop()
```
